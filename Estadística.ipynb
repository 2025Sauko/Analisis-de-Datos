```python
import statistics as st
```

**Media** = Valor promedio.

```python
# Valores para salario
muestra = [110, 30, 36, 47, 50, 52, 52, 56, 60, 63, 70, 70]
print("Media de la muestra: ", st.mean(muestra))
```

```python
# Cálculo de media
# 1. Calcular la suma de los valores de la lista
suma = sum(muestra)
print("Suma:" , suma)
# 2. Dividir la suma entre el número de elementos de la lista
print("Número de elementos de la muestra: ", len(muestra))
media = suma / len(muestra)
print("Media de la muestra:", media)
```

**Mediana** = Valor central.

```python
print("Mediana de la muestra:", st.median(muestra))
```

```python
# Cálculo de mediana
# 1. Ordenar valores ascendentemente
muestra.sort()
# 2. Obtener mediana
longitud=len(muestra)
print("La muestra ordenada es: ", muestra)
print("La longitud de la muestra es: ", longitud)
mitad = longitud // 2
print("La mitad de la muestra es: ", mitad)
if longitud % 2 == 0: # la longitud es par 12 % 2 == 0, el residuo de la división es 0
  print("Valores centrales: ", muestra[mitad-1], muestra[mitad])
  mediana = (muestra[mitad-1]+muestra[mitad])/2
else :
  mediana = muestra[mitad]
print("La mediana de la muestra es: ", mediana)
```

**Moda** = Valor más frecuente.

```python
# Valores para salario
muestra = [30, 36, 47, 50, 52, 52, 56, 60, 63, 70, 70, 110]
print("La moda de la muestra es:", st.multimode(muestra))
```

**Varianza** = Dispersión de los datos.

```python
import numpy as np
# Número de mascotas que cada persona tiene
datos = [0, 1, 5, 7, 9, 10, 14]
print("La varianza de la población es: ", np.var(datos)) # Obtiene e imprime la varianza de la población
```

```python
# Cálculo de la varianza de una población
def varianza_poblacion(valores):
  media = sum(valores) / len(valores)
  varianza = sum((v - media) ** 2 for v in valores) / len(valores)
  return varianza

print("La varianza de la población es:", varianza_poblacion(datos))
```

```python
print("La varianza de la muestra es: ", st.variance(datos)) # Obtiene e imprime varianza de la muestra
```

```python
# Cálculo de la varianza de una muestra
def varianza_muestra(valores):
  media = sum(valores) / len(valores)
  varianza = sum((v - media) ** 2 for v in valores) / (len(valores)-1)
  return varianza

print("La varianza de la muestra es:", varianza_muestra(datos))
```

**Desviación estándar** = Raíz cuadrada de la varianza.

```python
varianza = np.var(datos)
print("La desviación estándar de la población es: ", np.sqrt(varianza))
```

```python
# Cálculo de la desviación estándar de una población
from math import sqrt
def desv_est_poblacion(valores):
  return sqrt(varianza_poblacion(valores))

print("La desviación estándar de la población es:", desv_est_poblacion(datos))
```

```python
varianza = st.variance(datos)
print("La desviación estándar de la muestra es: ", sqrt(varianza))
```

```python
# Cálculo de la desviación estándar de una muestra
def desv_est_muestra(valores):
  return sqrt(varianza_muestra(valores))

print("La desviación estándar de la muestra es:", desv_est_muestra(datos))
```
