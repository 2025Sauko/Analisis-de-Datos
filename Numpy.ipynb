```python
# Ejemplo 1. Conversi贸n de temperaturas Celsius a Fahrenheit
# Antes de poder utilizar Numpy primero se tiene que importar
import numpy as np
```

```python
# Se declara una lista con los valores de las temperaturas
valores_celsius = [20.1, 20.8, 21.9, 22.5, 22.7, 22.3, 21.8, 21.2, 20.9, 20.1]
```

```python
# Se convierte la lista valores_celsius en un arreglo de una dimensi贸n de Numpy
C = np.array(valores_celsius)
print(C)
```

```python
type(C)
```

```python
# Se realiza la conversi贸n y se imprime
print(C * 9 / 5 + 32)
```

```python
# Esta es la soluci贸n utilzando listas
valores_fahrenheit = [x * 9 / 5 + 32 for x in valores_celsius ]
print(valores_fahrenheit)
```
